```python
numbers = []

for i in range(1, 101):
    if i % 15 == 0:
        numbers.append('FizzBuzz')
    if i % 5 == 0:
        numbers.append('Buzz')
    if i % 3 == 0:
        numbers.append('Fizz')
    else:
        numbers.append(i)
numbers
```
